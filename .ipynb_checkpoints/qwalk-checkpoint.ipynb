# Criando Grafo Cn
Abaixo, criando um grafo ciclo, assim como uma visualização

```python
Cn = 100
def create_graph(Cn,plot=False):  
    G = nx.cycle_graph(Cn)
    if plot:
        nx.draw_kamada_kawai(G, node_size=50)
        plt.show()
    return G
        
G = create_graph(Cn,plot=True)
```

Inicializando o vetor do estado inicial

```python
gamma = 1
adj = nx.to_scipy_sparse_matrix(G)
graph_size = Cn
initial_state = np.zeros(graph_size ,dtype=complex)
initial_state[0] = 1/math.sqrt(2)
initial_state[1] = 1/math.sqrt(2)


```

Realizando a primeira exponenciação

```python
def exp_matrix(adj,gamma,t):
    return expm(1j*gamma*adj*t)

def plot_dist(adj,gamma,t,initial_state):
    adjnew = exp_matrix(adj,gamma,t=t)
    initial_state = adjnew.dot(initial_state)
    prob = initial_state
    for i in range(len(initial_state)):
        prob[i] = pow(abs(initial_state[i]),2)
    plt.bar([x for x in range(0,Cn)],prob)
    plt.show()

# for t in range(0,1000,50):
plot_dist(adj,15,gamma,initial_state)
```
